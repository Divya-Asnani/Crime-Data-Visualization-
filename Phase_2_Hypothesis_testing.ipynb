```python
import numpy as np; np.set_printoptions(precision=20, suppress=False)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency, f_oneway

# base copy
df_h = df.copy()

# clean gender
df_h['Vict Sex'] = df_h['Vict Sex'].astype(str).str.strip().str.upper()
df_h['Vict Sex'] = df_h['Vict Sex'].replace(['X','H','-'], 'Unknown')

# clean age
df_h['Vict Age'] = pd.to_numeric(df_h['Vict Age'], errors='coerce')
df_h = df_h[(df_h['Vict Age'] > 0) & (df_h['Vict Age'] <= 100)]

# hour
df_h['hour'] = df_h['TIME OCC'] // 100

# weapon flag
df_h['Weapon Used'] = (df_h['Weapon Desc'] != 'Unknown').astype(int)

# reduce classes for stability
top_crimes = df_h['Crm Cd Desc'].value_counts().head(8).index
df_h = df_h[df_h['Crm Cd Desc'].isin(top_crimes)]
```

```python
#hypothesis 1 gender vs crime type
# H0 (Null Hypothesis):
# Victim gender and crime type are independent (no relationship)

# H1 (Alternative Hypothesis):
# Victim gender and crime type are dependent (there is a relationship)

df_h1 = df_h[['Vict Sex', 'Crm Cd Desc']].copy()

# Use only M/F for clean test
df_h1 = df_h1[df_h1['Vict Sex'].isin(['M','F'])]

ct1 = pd.crosstab(df_h1['Vict Sex'], df_h1['Crm Cd Desc'])

chi2_1, p_1, dof_1, exp_1 = chi2_contingency(ct1)

print("H1 Chi-square:", chi2_1)
print("p-value:", f"{p_1:.20e}")

# For interpretation (proportions)
ct1_norm = ct1.div(ct1.sum(axis=1), axis=0)
print(ct1_norm)

```

```python
# H0 (Null Hypothesis):
# Mean victim age is the same across all crime types

# H1 (Alternative Hypothesis):
# At least one crime type has a different mean victim age

df_h2 = df_h[['Vict Age', 'Crm Cd Desc']].copy()

groups = []
labels = []

for crime in df_h2['Crm Cd Desc'].unique():
    ages = df_h2[df_h2['Crm Cd Desc'] == crime]['Vict Age']
    if len(ages) > 50:
        groups.append(ages)
        labels.append(crime)

f_stat, p_2 = f_oneway(*groups)

print("H2 F-statistic:", f_stat)
print("H2 p-value:", p_2)

# Mean age per crime (for interpretation)
mean_age = df_h2.groupby('Crm Cd Desc')['Vict Age'].mean().sort_values()
print(mean_age)
```

```python
# H0 (Null Hypothesis):
# Crime type is independent of whether the incident occurred on a weekend

# H1 (Alternative Hypothesis):
# Crime type depends on whether the incident occurred on a weekend

df_h3 = df_h[['Is_Weekend', 'Crm Cd Desc']].copy()

ct3 = pd.crosstab(df_h3['Is_Weekend'], df_h3['Crm Cd Desc'])

chi2_3, p_3, dof_3, exp_3 = chi2_contingency(ct3)

print("H3 Chi-square:", chi2_3)
print("p-value:", f"{p_3:.20e}")

# Proportions
ct3_norm = ct3.div(ct3.sum(axis=1), axis=0)
print(ct3_norm)
```

```python
# H0:
# Crime type is independent of premise category

# H1:
# Crime type depends on premise category

df_h4 = df_h[['Premise_Category', 'Crm Cd Desc']].copy()

ct4 = pd.crosstab(df_h4['Premise_Category'], df_h4['Crm Cd Desc'])

chi2_4, p_4, _, _ = chi2_contingency(ct4)

print("H4 p-value:", p_4)

ct4_norm = ct4.div(ct4.sum(axis=1), axis=0)
print(ct4_norm)
```
